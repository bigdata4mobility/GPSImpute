```python
class HolidayResults:
    def __init__(self, results_files, skmob_files_path):
        self.results_files = results_files
        self.skmob_files_path = skmob_files_path

    def classicalMetrics(self):
        results_ids = [f.split('_')[2].split('.')[0] for f in self.results_files]

        self.orig_mae = np.array([])
        self.orig_rmse = np.array([])
        self.orig_mad = np.array([])
        self.orig_MAPE = np.array([])
        self.orig_MAXAPE = np.array([])
        self.orig_TAPE = np.array([])
        self.orig_PCM = np.array([])
        self.orig_DF = np.array([])
        self.orig_AREA = np.array([])
        self.orig_CL = np.array([])
        self.orig_DTW = np.array([])
        self.orig_MSE = np.array([])
        self.mod_mae = np.array([])
        self.mod_rmse = np.array([])
        self.mod_mad = np.array([])
        self.mod_MAPE = np.array([])
        self.mod_MAXAPE = np.array([])
        self.mod_TAPE = np.array([])
        self.mod_PCM = np.array([])
        self.mod_DF = np.array([])
        self.mod_AREA = np.array([])
        self.mod_CL = np.array([])
        self.mod_DTW = np.array([])
        self.mod_MSE = np.array([])

        for id in set(results_ids):
            result_file = "/Users/ekinokos2/Library/CloudStorage/OneDrive-UW/GPSImputeOct2023/results/holidays/all_results/results_{}.csv".format(id)

            res = pd.read_csv(result_file, header=0)

            self.orig_mae = np.append(self.orig_mae, res['MAE'][0])
            self.orig_rmse = np.append(self.orig_rmse, res['RMSE'][0])
            self.orig_mad = np.append(self.orig_mad, res['MAD'][0])
            self.orig_MAPE = np.append(self.orig_MAPE, res['MAPE'][0])
            self.orig_MAXAPE = np.append(self.orig_MAXAPE, res['MAXAPE'][0])
            self.orig_TAPE = np.append(self.orig_TAPE, res['TAPE'][0])
            self.orig_PCM = np.append(self.orig_PCM, res['PCM'][0])
            self.orig_DF = np.append(self.orig_DF, res['DF'][0])
            self.orig_AREA = np.append(self.orig_AREA, res['AREA'][0])
            self.orig_CL = np.append(self.orig_CL, res['CL'][0])
            self.orig_DTW = np.append(self.orig_DTW, res['DTW'][0])
            self.orig_MSE = np.append(self.orig_MSE, res['MSE'][0])
            self.mod_mae = np.append(self.mod_mae, res['MAE'][1])
            self.mod_rmse = np.append(self.mod_rmse, res['RMSE'][1])
            self.mod_mad = np.append(self.mod_mad, res['MAD'][1])
            self.mod_MAPE = np.append(self.mod_MAPE, res['MAPE'][1])
            self.mod_MAXAPE = np.append(self.mod_MAXAPE, res['MAXAPE'][1])
            self.mod_TAPE = np.append(self.mod_TAPE, res['TAPE'][1])
            self.mod_PCM = np.append(self.mod_PCM, res['PCM'][1])
            self.mod_DF = np.append(self.mod_DF, res['DF'][1])
            self.mod_AREA = np.append(self.mod_AREA, res['AREA'][1])
            self.mod_CL = np.append(self.mod_CL, res['CL'][1])
            self.mod_DTW = np.append(self.mod_DTW, res['DTW'][1])
            self.mod_MSE = np.append(self.mod_MSE, res['MSE'][1])

    def skmobMetrics(self):
        os.chdir(self.skmob_files_path)
        self.orig_no_loc_error = np.array([])
        self.orig_rg_error = np.array([])
        self.orig_k_rg_error = np.array([])
        self.orig_spat_burst_error = np.array([])
        self.orig_rand_entr_error = np.array([])
        self.orig_real_entr_error = np.array([])
        self.orig_uncorr_entr_error = np.array([])
        self.orig_max_dist_error = np.array([])
        self.orig_dist_straight_error = np.array([])
        self.orig_max_dist_home_error = np.array([])
        self.orig_rec_error = np.array([])
        self.orig_freq_error = np.array([])
        self.mod_no_loc_error = np.array([])
        self.mod_rg_error = np.array([])
        self.mod_k_rg_error = np.array([])
        self.mod_spat_burst_error = np.array([])
        self.mod_rand_entr_error = np.array([])
        self.mod_real_entr_error = np.array([])
        self.mod_uncorr_entr_error = np.array([])
        self.mod_max_dist_error = np.array([])
        self.mod_dist_straight_error = np.array([])
        self.mod_max_dist_home_error = np.array([])
        self.mod_rec_error = np.array([])
        self.mod_freq_error = np.array([])
        
        

        for file in glob.glob("*.csv"):
            res = pd.read_csv(file, header=0)
            # Store each metric in a numpy array
            self.orig_no_loc_error = np.append(self.orig_no_loc_error, res['no_loc_error'][0])
            self.orig_rg_error = np.append(self.orig_rg_error, res['rg_error'][0])
            self.orig_k_rg_error = np.append(self.orig_k_rg_error, res['k_rg_error'][0])
            self.orig_spat_burst_error = np.append(self.orig_spat_burst_error, res['spat_burst_error'][0])
            self.orig_rand_entr_error = np.append(self.orig_rand_entr_error, res['rand_entr_error'][0])
            self.orig_real_entr_error = np.append(self.orig_real_entr_error, res['real_entr_error'][0])
            self.orig_uncorr_entr_error = np.append(self.orig_uncorr_entr_error, res['uncorr_entr_error'][0])
            self.orig_max_dist_error = np.append(self.orig_max_dist_error, res['max_dist_error'][0])
            self.orig_dist_straight_error = np.append(self.orig_dist_straight_error, res['dist_straight_error'][0])
            self.orig_max_dist_home_error = np.append(self.orig_max_dist_home_error, res['max_dist_home_error'][0])
            self.orig_rec_error = np.append(self.orig_rec_error, res['recency'][0])
            self.orig_freq_error = np.append(self.orig_freq_error, res['freq_rank'][0])
            self.mod_no_loc_error = np.append(self.mod_no_loc_error, res['no_loc_error'][1])
            self.mod_rg_error = np.append(self.mod_rg_error, res['rg_error'][1])
            self.mod_k_rg_error = np.append(self.mod_k_rg_error, res['k_rg_error'][1])
            self.mod_spat_burst_error = np.append(self.mod_spat_burst_error, res['spat_burst_error'][1])
            self.mod_rand_entr_error = np.append(self.mod_rand_entr_error, res['rand_entr_error'][1])
            self.mod_real_entr_error = np.append(self.mod_real_entr_error, res['real_entr_error'][1])
            self.mod_uncorr_entr_error = np.append(self.mod_uncorr_entr_error, res['uncorr_entr_error'][1])
            self.mod_max_dist_error = np.append(self.mod_max_dist_error, res['max_dist_error'][1])
            self.mod_dist_straight_error = np.append(self.mod_dist_straight_error, res['dist_straight_error'][1])
            self.mod_max_dist_home_error = np.append(self.mod_max_dist_home_error, res['max_dist_home_error'][1])
            self.mod_rec_error = np.append(self.mod_rec_error, res['recency'][1])
            self.mod_freq_error = np.append(self.mod_freq_error, res['freq_rank'][1])
```

```python
# Plot
def plotClassicalMetrics(metrics, **kwargs):
    plt.figure(figsize=(15, 5))
     # Increase x-axis tick font size
    plt.rc('xtick', labelsize=12)
    # Increase y-axis tick font size
    plt.rc('ytick', labelsize=12)
   
    plt.subplot(2, 3, 1)

    plt.boxplot([metrics.orig_mae, metrics.mod_mae])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('MAE')

    plt.subplot(2, 3, 2)
    plt.boxplot([metrics.orig_rmse, metrics.mod_rmse])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('RMSE')

    plt.subplot(2, 3, 3)
    plt.boxplot([metrics.orig_mad, metrics.mod_mad])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('MAD')

    plt.subplot(2, 3, 4)
    plt.boxplot([metrics.orig_MAPE, metrics.mod_MAPE])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('MAPE')
    plt.ylim(-10, 50)

    plt.subplot(2, 3, 5)
    plt.boxplot([metrics.orig_MAXAPE, metrics.mod_MAXAPE])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('MAXAPE')
    plt.ylim(-100,2000)

    plt.subplot(2, 3, 6)
    plt.boxplot([metrics.orig_TAPE, metrics.mod_TAPE])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('TAPE')
    plt.ylim(-100, 5000)

    plt.tight_layout()

    if kwargs:
            plt.ylim(kwargs['ylim'])
        
    plt.show()

plotClassicalMetrics(metrics)
```

```python
def plotSkmobMetrics(metrics, **kwargs):
    plt.figure(figsize=(15, 5))
     # Increase x-axis tick font size
    plt.rc('xtick', labelsize=12)
    
    # Increase y-axis tick font size
    plt.rc('ytick', labelsize=12)
   
    plt.subplot(2, 3, 1)
    plt.boxplot([metrics.orig_no_loc_error, metrics.mod_no_loc_error])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('Number of locations')
    plt.ylim(-10, 20)

    plt.subplot(2, 3, 2)
    plt.boxplot([metrics.orig_rg_error, metrics.mod_rg_error])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('Radius of gyration')

    plt.subplot(2, 3, 3)
    plt.boxplot([metrics.orig_dist_straight_error, metrics.mod_dist_straight_error])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('Straight line travel distance')

    # Ignore NaNs
    plt.subplot(2, 3, 4)
    plt.boxplot([metrics.orig_spat_burst_error[~np.isnan(metrics.orig_spat_burst_error)], metrics.mod_spat_burst_error[~np.isnan(metrics.mod_spat_burst_error)]])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('Spatial burstiness')

    plt.subplot(2, 3, 5)
    plt.boxplot([metrics.orig_rand_entr_error, metrics.mod_rand_entr_error])
    plt.xticks([1, 2], ['Holiday Model', 'Regular Model'])
    plt.title('Random entropy')
    plt.ylim(-1, 1.0)

    plt.subplot(2, 3, 6)
    plt.boxplot([metrics.orig_real_entr_error, metrics.mod_real_entr_error])
    plt.xticks([1, 2] , ['Holiday Model', 'Regular Model'])
    plt.title('Real entropy')
    plt.ylim(-2, 2)

    plt.tight_layout()

    if kwargs:
        plt.ylim(kwargs['ylim'])

    
    plt.show()

plotSkmobMetrics(metrics)
```
